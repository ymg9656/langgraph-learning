## 개념 요약

### 프롬프트의 4가지 구성 요소

| 구성 요소 | 설명 | 예시 |
|-----------|------|------|
| **역할(Role)** | LLM에게 전문가 역할 부여 | "당신은 시니어 QA 엔지니어입니다" |
| **맥락(Context)** | 배경 정보, 도메인 지식 | "Spring Boot 기반 REST API 프로젝트입니다" |
| **지시(Instruction)** | 구체적으로 무엇을 해야 하는지 | "코드의 버그를 찾아 수정 방안을 제시하세요" |
| **출력 형식(Format)** | 원하는 응답 형태 | "JSON 형식으로, bugs와 fixes 키를 포함하세요" |

### System vs Human 메시지

- **System**: 역할, 규칙, 금지 사항, 출력 형식 (거의 변경되지 않음)
- **Human**: 입력 데이터, 변수, 구체적 요청문 (매 요청마다 다름)

### ChatPromptTemplate 핵심 패턴

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "역할 + 규칙"),
    ("human", "데이터: {variable}"),  # {변수}는 나중에 치환
])
messages = prompt.format_messages(variable="실제값")
```
