## 개념 요약

### AI 에이전트란?
AI 에이전트는 **자율적으로 목표를 달성하기 위해 도구를 사용하는 AI 시스템**입니다.

| 구분 | 챗봇 | 에이전트 |
|------|------|----------|
| 동작 방식 | 단일 턴 응답 | 다단계 계획·실행 |
| 도구 사용 | 없음 | 검색, 코드 실행 등 |
| 자율성 | 낮음 | 높음 |

### 에이전트 5대 구성 요소
1. **인지 (Perception)**: 환경으로부터 정보를 수집
2. **추론 (Reasoning)**: LLM이 다음 행동을 결정
3. **행동 (Action)**: 도구를 호출하거나 응답 생성
4. **반복 (Loop)**: 목표 달성까지 인지→추론→행동 반복
5. **도구 (Tools)**: 외부 기능(검색, 계산, DB 등) 활용

### 에이전트가 부적합한 경우
- 단순한 단일 응답으로 충분한 경우
- 결정론적(deterministic) 처리가 필요한 경우
- 지연 시간이 극도로 짧아야 하는 경우
- 예측 불가능한 행동이 허용되지 않는 경우

## 에이전트 아키텍처 패턴 3가지

에이전트를 설계할 때 대표적으로 쓰이는 3가지 패턴이 있습니다.

- **ReAct**: 추론(Reasoning) + 행동(Acting) 반복. "생각하고 행동한다"
  - LLM이 생각(Thought) → 행동(Action) → 관찰(Observation) 사이클을 반복하며 목표에 도달합니다.
  - 적합한 상황: 검색·분석을 반복하며 정보를 점진적으로 수집해야 할 때

- **Plan-and-Execute**: 먼저 전체 계획을 세우고, 단계별로 실행
  - 플래너(Planner) LLM이 작업 목록을 한 번에 생성하고, 실행자(Executor)가 순서대로 처리합니다.
  - 적합한 상황: 작업 순서가 사전에 명확하고, 단계가 많은 장기 태스크

- **Tool-use**: 상황에 따라 적절한 도구를 선택하여 사용
  - LLM이 여러 도구(Tool) 중 문맥에 맞는 것을 선택해 호출합니다.
  - 적합한 상황: 계산기, 번역기, 날씨 API 등 이종(異種) 도구를 상황별로 골라 써야 할 때

```python
# TODO: 각 시나리오에 적합한 아키텍처 패턴을 선택하세요
# 선택지: "ReAct", "Plan-and-Execute", "Tool-use"

q6_answer = ""  # 검색 → 결과 분석 → 추가 검색이 필요한 에이전트
q7_answer = ""  # 10단계 보고서를 순서대로 작성하는 에이전트
q8_answer = ""  # 계산기, 번역기, 날씨API 중 골라 사용하는 에이전트

print(f"Q6: {q6_answer}")
print(f"Q7: {q7_answer}")
print(f"Q8: {q8_answer}")
```
