## 개념 요약

**동기 실행**: 한 작업이 끝날 때까지 다음 작업이 기다림
```
Task 1: [████████████] 완료
Task 2:               [████████████] 완료
총 시간: 작업1 + 작업2
```

**비동기 실행**: 여러 작업이 동시에 진행됨
```
Task 1: [████████████] 완료
Task 2: [████████████] 완료 (동시 시작!)
총 시간: max(작업1, 작업2)
```

핵심 패턴:
```python
async def my_func():      # 비동기 함수 정의
    await asyncio.sleep(1)  # 비동기 대기
    return "완료"

# 동시 실행
results = await asyncio.gather(my_func(), my_func())
```
