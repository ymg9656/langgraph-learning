## 개념 요약

### 조건부 엣지 (Conditional Edges)

상태에 따라 다른 노드로 분기할 때 사용합니다.

```
                    ┌─ score >= 80 ──→ [congratulate] ─┐
START → [check_score]
                    └─ score < 80  ──→ [encourage]    ─┘→ END
```

### 핵심 패턴

```python
def routing_function(state: MyState) -> str:
    if state["score"] >= 80:
        return "pass"  # 경로 이름
    else:
        return "fail"

graph.add_conditional_edges(
    "check_score",          # 출발 노드
    routing_function,        # 라우팅 함수
    {                        # 경로명 → 노드명 매핑
        "pass": "congratulate",
        "fail": "encourage"
    }
)
```
